###Import Data

```python
%matplotlib inline
from matplotlib import pyplot
import pickle
import pandas as pd
import numpy as np
import datetime as dt

with open("movies_data.pkl", 'r') as picklefile: 
    movies_scraped, movies_skipped = pickle.load(picklefile)

full_df = pd.DataFrame(movies_scraped)
full_df.dtypes
```

### Define a initial subset of relevant variables

```python
full_df = full_df[['movie_title', 'genre', 'budget', 'opening_income_wend', 'opening_theaters', 'release_date', 'runtime_mins']]
```

### Drop rows with NaN values

```python
print 'rows with NaN', len(full_df)
df = full_df.dropna()
print 'rows without NaN', len(df)
```

### Bin 'release_date' in the relative weekend of the year (aaaaww)

```python
df['release_date'] = df['release_date'].astype(dt.date)
df.set_index('release_date', inplace=True)
df['release_weekend'] = df.index.year * 100 + df.index.week    #check first day of the week
df.reset_index(inplace=True)
df = df[['movie_title', 'genre', 'budget', 'opening_income_wend', 'opening_theaters', 'release_weekend', 'runtime_mins']]
df.head()
```

### Add the variable 'opening_count' (number of movies that are released in the same weekend)

```python
comp_df = full_df[['movie_title', 'opening_income_wend', 'opening_theaters', 'release_date']]
comp_df = comp_df.dropna()

comp_df['release_date'] = comp_df['release_date'].astype(dt.date)
comp_df.set_index('release_date', inplace=True)
comp_df['release_weekend'] = comp_df.index.year * 100 + comp_df.index.week    #check first day of the week
comp_df.reset_index(inplace=True)
comp_df = comp_df[['movie_title', 'opening_income_wend', 'opening_theaters', 'release_weekend']]


op_week_count = comp_df[['movie_title','release_weekend']].groupby('release_weekend',as_index=False).count()
op_week_count.columns = ['release_weekend','opening_count']
op_week_count.sort('opening_count', ascending=False).head()
```

```python
merged_df = pd.merge(df, op_week_count, on='release_weekend')
merged_df.head(8)
```

### First linear regression: opening income vs (budget, opening_theaters)

```python
import statsmodels.formula.api as sm
merged_df['ones'] = 1.0
Y = merged_df['opening_income_wend']
X = merged_df[['budget', 'opening_theaters', 'ones']]
```

```python
linmodel = sm.OLS(Y,X).fit()
linmodel.summary()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(18,8))
plt.scatter(merged_df.budget, merged_df.opening_income_wend)
plt.title('How do budget and number of opening theaters relate to opening income in the first weekend?')
plt.xlabel('Budget (hundred million dollars)')
plt.ylabel('Box office opening weekend (hundred million dollars)')
plt.ylim([0,200000000])
plt.xlim([0,300000000])
plt.plot(merged_df.budget, linmodel.predict(X), 'rs')
```

### Evaluate the model with training and testing set

```python
import random

#Sample for training
train_index = random.sample(merged_df.index, int(0.75*len(merged_df)))
X_train = merged_df[['budget', 'opening_theaters', 'ones']].ix[train_index]
Y_train = merged_df['opening_income_wend'].ix[train_index]

#fitting the model
linmodel = sm.OLS(Y_train,X_train).fit()
linmodel.summary()
```

```python
#plotting
plt.figure(figsize=(18,8))
plt.scatter(merged_df.budget, merged_df.opening_income_wend)
plt.title('How do budget and number of opening theaters relate to opening income in the first weekend? (trainimg set)')
plt.xlabel('Budget (hundred million dollars)')
plt.ylabel('Box office opening weekend (hundred million dollars)')
plt.ylim([0,200000000])
plt.xlim([0,300000000])
plt.plot(merged_df.budget.ix[train_index], linmodel.predict(X_train), 'rs')
```

```python
#Test sample
X_test = merged_df[['budget', 'opening_theaters', 'ones']].drop(train_index)
Y_test = merged_df['opening_income_wend'].drop(train_index)

#plotting
plt.figure(figsize=(18,8))
plt.plot(Y_test, linmodel.predict(X_test), 'go')
plt.xlabel('Actual opening income')
plt.ylabel('Predicted opening income By Model')
plt.title('How do budget and number of opening theaters relate to opening income in the first weekend? (testing set)')
plt.plot(Y_test,Y_test,'m--', label='Ideal Prediction')
plt.legend()
```

```python
import math
Y_pred_test = linmodel.predict(X_test)
print 'Mean square error = %.2f' % math.sqrt(np.mean((Y_pred_test - Y_test)**2))
```
