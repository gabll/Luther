### <a id='Import Data'></a> Import data

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pickle
import pandas as pd
import numpy as np
import datetime as dt
import BOM_data_utils


with open("movies_data.pkl", 'r') as picklefile: 
    movies_scraped, movies_skipped = pickle.load(picklefile)

pickle_df = pd.DataFrame(movies_scraped)
pickle_df['release_date'] = pickle_df['release_date'].astype(dt.date)
pickle_df.dtypes
```

### <a id='Compute release weekend and season'></a> Compute release weekend and season

```python
date_df = pickle_df
date_df.set_index('release_date', inplace=True)
date_df['release_weekend'] = date_df.index.year * 100 + date_df.index.week
date_df.reset_index(inplace=True)
date_df['release_season'] = pd.Series([BOM_data_utils.get_season(i.date()) for i in date_df.release_date])
date_df.head(10)
```

### <a id='Bin genres in a smaller set'></a> Bin genres in a smaller set

```python
genre_df = date_df
explore_BOM_genres = genre_df[['movie_title','genre']].groupby('genre').count()
explore_BOM_genres.columns = ['movie_count']
print explore_BOM_genres.head(10)
print explore_BOM_genres.describe()
```

```python
def bin_genre(BOM_genre):
    default_genres = ['adventure',
                        'animation',
                        'action',
                        'comedy',
                        'drama',
                        'sci-fi',
                        'horror',
                        'romant',
                        'documentary',
                        'musical',
                        'crime',
                        'legal',
                        'thriller',
                        'fantasy',
                        'war',
                        'western']
    if BOM_genre:
        for genre in default_genres:
            if genre in BOM_genre.lower():
                return genre
    else:
        return None

genre_df['genre'] = pd.Series([bin_genre(i) for i in genre_df.genre])
genre_df['genre'] = genre_df['genre'].replace('crime','thriller')
explore_BOM_genres = genre_df[['movie_title','genre']].groupby('genre').count()
explore_BOM_genres.columns = ['movie_count']
explore_BOM_genres
```

### <a id='Group MPAA rating'></a> Group MPAA rating

```python
rating_df = genre_df
genre_df['rating'] = genre_df['rating'].replace('PG-13','PG')
genre_df['rating'] = genre_df['rating'].replace('NC-17','R')
genre_df['rating'] = genre_df['rating'].replace('Unrated','None')
rating_df.head()
```

### <a id='Drop not interesting variables'></a> Drop not interesting variables

```python
final_df = rating_df
final_df = final_df.drop(['actors', 'director', 'distributor'],1)
final_df.head(10)
```

### <a id='Save to a pickle file'></a> Save to a pickle file

```python
import pickle
with open('BOM_dataframe.pkl', 'w') as picklefile:    
    pickle.dump(final_df, picklefile)
```
